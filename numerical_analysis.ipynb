```python
# 이분법 근찾기
import math

def func(x):   # 근을 찾을 함수 설정
    return math.log(x+5.0) + x


xh = 3.0
xl = -3.0
e= 0.000001
i = 1

print("xh = ", xh)
print("xl = ", xl)
print("e = ", e)

while (xh - xl) > e:

    xm = (xh + xl) / 2.0
    
    if (func(xm)*func(xh)) == float(0):
        break
    elif(func(xm)*func(xl) < float(0)):
        xh = xm
    elif (func(xm)*func(xh)) < float(0):
        xl = xm
    else:
        print("somthing wrong")
        break
    
    print("[iteration %d] The root is %.8f error is %.8f"%(i ,xm, xh-xl))   # 결과출력
    i+=1  
```

```python
# 이분법 극소값 찾기

def func(x):   #  극솟값을 찾을 함수 설정
    return 2.5*x*x-7.0*x+2.5

xl = -3.0
xm = 0.0
xr = 0.0
s = 2.0
e= 0.001
i = 1

print("xl= ", xl)
print("s = ", s)
print("e = ", e)

while s > e: # 간격이 오차범위보다 작을 때까지 반복

    xm = xl + s
    xr = xm + s # 세 점을 시작점 부터 같은 간격으로 배치
    
    if (func(xm) < func(xl)) and (func(xm) < func(xr)): # 극솟값이 범위내 있다고 판단되면 간격을 줄임
        s = s/2.0
    else:
        xl = xm # 범위내 없을 경우 간격만큼 이동함
    
    print("[iteration %d] The local minimum is (%.8f, %.8f) error is %.8f"%(i ,xm, func(xm), s))   # 결과출력
    i+=1  
```

```python
# 영점 곡선 추적 알고리즘
import math
import matplotlib.pyplot as plt

def func(x, y):   #  함수 설정
    return x**2 + y**2 - 4

def root(x, y, p, q, e):  # 이분법을 이용해 근을 찾는다.   x,y는 시작점 p,q는 이동할수있는 (x,y)값을 의미
    while True:
        xn = x + p
        yn = y + q

        if (func(x,y)*func(xn,yn)) <= 0.0 : # 성립할 경우 사이에 근이 있다는 뜻
            if (math.fabs(p) > e) or (math.fabs(q) > e): # 오차범위안에 안들 경우 더하는 값을 절반으로 줄임.
                p = p / 2.0
                q = q / 2.0
            else:
                return xn, yn
        else:
            x = xn
            y = yn

i = 1
x = y = u = v = d = A = B = xi = yi = 0.0

w = 0.1
c = 0.1
e = 0.000001

u,v = root(xi, yi, c, 0.0, e)
x,y = root(xi, yi+w, c, 0.0, e) #  이분격자법을 이용해 해가되는 두 점을 찾아낸다.

print("x\t\ty\t\tf(x,y)")

while True:
    d = math.sqrt((x-u)**2 + (y-v)**2)
    A = (x-u) / d  # cos
    B = (y-v) / d  # sin
    u = x
    v = y
    x += math.fabs(w)*(A-B)
    y += math.fabs(w)*(A+B)
    
    xi, yi = root(x, y, c*B, -c*A, e)
    
    
    print("%.8f\t%.8f\t%.8f"%(xi, yi, func(xi,yi)))
    
    x = xi
    y = yi
    plt.scatter(xi, yi, s=6)   # 사이즈 6의 점으로 좌표평면상에 찍음.
    plt.axis([-2.5, 2.5, -2.5, 2.5]) # 좌표의 크기설정
    
    i += 1
    
    if i==500:  # iteration이 500회에 도달했을때, 좌표찍힌 것들 출력 후 종료.
        plt.show()
        break
```

```python
# 수치적분 by 사다리꼴공식(trapezoidal rule of integration)

import numpy as np

def function(t):  # 적분할 함수
    return 2000 * np.log(140000/(140000-2100*t)) - 9.8 * t   


def Traplezoidal_rule(a, b, n, True_value):  # 사다리꼴 공식을 활용한 수치적분  a: front  b: back  n: divide number  True_value:적분정답값
    h = (b-a)/n
    
    middle = 0 
    for i in range(n-1):
        middle += function(a+(i+1)*h) # h단위로 쪼갰을 때,적분 중간지점 해당하는 y값 축적
    
    Approximate_Value = (b-a)/(2*n) * (function(a)+ 2*middle + function(b))
    
    Error = True_value - Approximate_Value
    Error_rate = abs(Error / True_value) * 100
    
    return Approximate_Value, Error, Error_rate



for i in range(1,9):
    Approximate_Value, Error, Error_rate = Traplezoidal_rule(8, 30, i, 11061)
    
    print("n: %d \t Value: %.2f \t Error: %.2f  \t Error_rate: %.4f"%(i, Approximate_Value, Error, Error_rate))
    
```

```python
# 수치적분 by multiple segment simpson's 1/3rd Rule

import numpy as np

def function(t):  # 적분할 함수
    return 2000 * np.log(140000/(140000-2100*t)) - 9.8 * t   

def x_n(n, a, b, h):
    return a + n * h
    

def Simpsons_rule(a, b, n, True_value):  #  a: front  b: back  n: divide number  True_value:적분정답값
    h = (b-a)/n
    
    middle_1 = 0
     
    for i in range(1, n):
        if i % 2 == 1:
            middle_1 += function(x_n(i,a,b,h)) # h단위로 쪼갰을 때,적분 중간지점 해당하는 y값 축적
            
    middle_2 = 0
    
    for i in range(2, n-1):
        if i % 2 == 0:
            middle_2 += function(x_n(i,a,b,h)) # h단위로 쪼갰을 때,적분 중간지점 해당하는 y값 축적
            
            
    Approximate_Value = (b-a)/(3*n) * (function(a)+ 4*middle_1 + 2*middle_2 + function(b))
    
    Error = True_value - Approximate_Value
    Error_rate = abs(Error / True_value) * 100
    
    return Approximate_Value, Error, Error_rate

for i in range(2,11,2):
    Approximate_Value, Error, Error_rate = Simpsons_rule(8, 30, i, 11061.34)
    
    print("n: %d \t Value: %.2f \t Error: %.2f  \t Error_rate: %.4f"%(i, Approximate_Value, Error, Error_rate))
```
